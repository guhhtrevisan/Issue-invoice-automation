### Passo a passo
1. Acessar o site 'login.html'
2. Preencher o login: email e senha; clicar para prosseguir
3. Preencher os campos
4. Clicar em 'emitir nota'
5. Fazer o processo para todos os usuários do arquio Excel 'Notas a emitir'
    a. Tratar os dados para estar de acordo com os campos a serem preenchidos

```python
# 1. Acessando site

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.select import Select

servico = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=servico)
```

```python
import os

caminho = os.getcwd()
arquivo = caminho + "\login.html"
driver.get(arquivo)
print(arquivo)
```

```python
# 2. Preenchendo login
import time

email = 'gustavo_trevisan118@gmail.com'
senha = '1234Abc'

driver.find_element(By.XPATH, '/html/body/div/form/input[1]').send_keys(email)
driver.find_element(By.XPATH, '/html/body/div/form/input[2]').send_keys(senha)
time.sleep(0.5)
driver.find_element(By.XPATH, '/html/body/div/form/button').click()
```

```python
import pandas as pd

df_clientes = pd.read_excel(fr'{caminho}\NotasEmitir.xlsx')
display(df_clientes.head())
```

```python
for i, nome in enumerate(df_clientes['Cliente']):    
    # Nome
    nome = nome
    driver.find_element(By.ID, 'nome').clear()
    driver.find_element(By.ID, 'nome').send_keys(nome)
    
    time.sleep(1)
    
    # Endereço
    endereco = df_clientes['Endereço'][i]
    driver.find_element(By.NAME, 'endereco').clear()
    driver.find_element(By.NAME, 'endereco').send_keys(endereco)
    
    time.sleep(1)
    
    # Bairro
    bairro = df_clientes['Bairro'][i]
    driver.find_element(By.NAME, 'bairro').clear()
    driver.find_element(By.NAME, 'bairro').send_keys(bairro)
    
    time.sleep(1)
    
    # Município
    cidade = df_clientes['Municipio'][i]
    driver.find_element(By.NAME, 'municipio').clear()
    driver.find_element(By.NAME, 'municipio').send_keys(cidade)
    
    time.sleep(1)
    
    # CEP
    cep = int(df_clientes['CEP'][i])
    driver.find_element(By.NAME, 'cep').clear()
    driver.find_element(By.NAME, 'cep').send_keys(cep)
    
    time.sleep(1)
    
    # UF (drop down)
    UF = df_clientes['UF'][i]
    elemento = driver.find_element(By.TAG_NAME, 'select')
    elemento_select = Select(elemento)
    elemento_select.select_by_visible_text(UF)
    
    time.sleep(1)
    
    #CNPJ/CPF
    CNPJ_CPF = int(df_clientes['CPF/CNPJ'][i])
    driver.find_element(By.NAME, 'cnpj').clear()
    driver.find_element(By.NAME, 'cnpj').send_keys(CNPJ_CPF)
    
    time.sleep(1)
    
    # Inscrição estadual
    inscricao_estadual = int(df_clientes['Inscricao Estadual'][i])
    driver.find_element(By.NAME, 'inscricao').clear()
    driver.find_element(By.NAME, 'inscricao').send_keys(inscricao_estadual)
    
    time.sleep(1)
    
    # Descrição do produto/serviço
    descricao = df_clientes['Descrição'][i]
    driver.find_element(By.NAME, 'descricao').clear()
    driver.find_element(By.NAME, 'descricao').send_keys(descricao)
    
    time.sleep(1)
    
    # Quantidade
    quantidade = int(df_clientes['Quantidade'][i])
    driver.find_element(By.NAME, 'quantidade').clear()
    driver.find_element(By.NAME, 'quantidade').send_keys(quantidade)
    
    time.sleep(1)
    
    # Valor unitário
    valor_un = float(df_clientes['Valor Unitario'][i])
    driver.find_element(By.NAME, 'valor_unitario').clear()
    driver.find_element(By.NAME, 'valor_unitario').send_keys(valor_un)
    
    time.sleep(1)
    
    # Valor total
    valor_total = float(df_clientes['Valor Total'][i])
    driver.find_element(By.NAME, 'total').clear()
    driver.find_element(By.NAME, 'total').send_keys(valor_total)
    
    time.sleep(1)

    # 4. Emitir nota
    driver.find_element(By.CLASS_NAME, 'registerbtn').click()
    
print('Loop finalizado! Todos as notas fiscais foram emitidas!')
```
